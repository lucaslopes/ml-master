Tanto o algoritmo de Dijkstra quanto o de value iteration usam uma função recorrente, parecida; um termo representa um ganho (distância, em Dijkstra) pequeno, enquanto o outro representa um ganho (distância) maior. O algoritmo de Dijkstra possui um aspecto de recursão que o value iteration não possui, a não ser que o número de iterações seja definido a priori (geralmente não é o caso), entretanto, ambos comunicam a ideia de que o problema maior pode ser dividido em etapas menores, e que, resolver o problema para os "vizinhos" de um estado/nó nos ajuda a resolver o problema do estado/nó de interesse.

# Questão 2
## 1. + 2.
Não é necessário calcular os custos para cada possível combinação das ações em cada estado. Tudo que precisamos são duas tabelas; uma com as probabilidades de transição entre estados, dada cada uma das ações, e outra com os custos (ganhos) associados às essas transições. O processo de escolher a ação que esperamos resultar no maior ganho a longo prazo consiste em, fixando o estado de origem e uma ação, somar, para todos os estados, a probabilidade dele ir para esse estado, dada aquela ação (pode ser consultado direto pela tabela), multiplicado pela soma do ganho imediato (pode ser consultado na outra tabela) com o quanto se espera ganhar no estado de destino (isso foi obtido na iteração anterior). Ao se fazer isso para todas as ações, naquele mesmo estado de origem,  nós vamos ter, o ganho esperado, a longo prazo, ao se tomar aquela ação, quando se está naquele estado, para aquela iteração. Nós guardamos o maior desses valores e a ação que resultou nesse valor, para cada estado, em cada iteração. Há mais um detalhe importante: os ganhos futuros nem sempre são tão valiosos quanto os imediatos, e isso pode ser levado em conta, na função recorrente, ao se multiplicar o ganho esperado no futuro por um coeficiente de desconto, que pode variar entre 0 e 1. Botar esse coeficiente com valor 1 indica que o ganho esperado no futuro tem o mesmo valor que o ganho imediato, enquanto que, botar esse coeficiente com valor 0 indica que o ganho esperado no futuro não importa, e a função deixa de ser recorrente.

Se nós não tivéssemos o modelo pronto nós poderíamos aprender ao longo do tempo, usando reinforcement learning, o valor que cada ação nos proporcionaria, para cada estado. O método de reinforcement learning visto em sala foi o Q-learning, que consiste em, a cada etapa, escolher a ação que nós achamos que nos dará o maior ganho, naquele estado (exploitation), ou, com uma probabilidade ajustável, escolher uma ação aleatória para tomar (exploration). Após escolher essa ação nós iremos consultar a tabela para, com as probabilidades adequadas, dado o estado de origem e a ação escolhida, transicionar para outro estado. Feito isso nós podemos então atualizar o ganho esperado a longo prazo, dado aquele estado de origem, para aquela ação, somando o ganho imediato com o ganho máximo, a longo prazo, que podemos obter no novo estado, vamos chamar essa soma de V. Nós podemos refinar esse modelo de duas formas; multiplicando o ganho máximo a longo prazo que podemos obter no novo estado por um coeficiente de desconto, vamos chamar essa nova soma de V' e, nós podemos usar o valor que esperamos ganhar, a longo prazo, no estado de origem e incrementar a ele, alpha*(V' menos o ganho, a longo prazo, que esperamos ter no estado de origem), onde alpha é a taxa de aprendizado.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#vector where i-th entry = probability of i customers buying one product during the week
prob_n_clients = np.array([0.25, 0.6, 0.15])

#matrix with state transition probability given a state and an action - first dimension is original state, second is action, third is destionation state
P = np.ndarray(shape=(3,3,3), dtype=float)

P[0][0][0] = 1
P[0][0][1] = 0
P[0][0][2] = 0
P[0][1][0] = sum(prob_n_clients[1:2])
P[0][1][1] = prob_n_clients[0]
P[0][1][2] = 0
P[0][2][0] = prob_n_clients[2]
P[0][2][1] = prob_n_clients[1]
P[0][2][2] = prob_n_clients[0]

P[1][0][0] = sum(prob_n_clients[1:2])
P[1][0][1] = prob_n_clients[0]
P[1][0][2] = 0
P[1][1][0] = prob_n_clients[2]
P[1][1][1] = prob_n_clients[1]
P[1][1][2] = prob_n_clients[0]
P[1][2][0] = 0
P[1][2][1] = 0
P[1][2][2] = 0

P[2][0][0] = prob_n_clients[2]
P[2][0][1] = prob_n_clients[1]
P[2][0][2] = prob_n_clients[0]
P[2][1][0] = 0
P[2][1][1] = 0
P[2][1][2] = 0
P[2][2][0] = 0
P[2][2][1] = 0
P[2][2][2] = 0

#matrix with the rewards gained after transitioning to another state given the previous state and the action taken- first dimension is original state, second is action, third is destionation state
R = np.ndarray(shape=(3,3,3), dtype=float)

R[0][0][0] = 0
R[0][0][1] = 0
R[0][0][2] = 0
R[0][1][0] = 45
R[0][1][1] = -55
R[0][1][2] = 0
R[0][2][0] = 100
R[0][2][1] = 0
R[0][2][2] = -100

R[1][0][0] = 98
R[1][0][1] = -2
R[1][0][2] = -2
R[1][1][0] = 143
R[1][1][1] = 53
R[1][1][2] = -47
R[1][2][0] = -2
R[1][2][1] = -2
R[1][2][2] = -2

R[2][0][0] = 196
R[2][0][1] = 96
R[2][0][2] = -4
R[2][1][0] = -4
R[2][1][1] = -4
R[2][1][2] = -4
R[2][2][0] = -4
R[2][2][1] = -4
R[2][2][2] = -4
```

```python
def trainMDP(P, R, gamma, epochs, threshold):
  
  n_states = P.shape[0]
  n_actions = P.shape[1]
  
  #list that will keep track of the maximum reward possible to get for each state
  V = []
  
  #list that will keep track of the action that maximizes the reward for each state
  A = []
  
  for i in range(n_states):
    V.append([0])
    
  for i in range(n_states):
    A.append([0])
    
  V_old = np.zeros([1, n_states])
  V_new = np.zeros([1, n_states])
  
  for e in range(epochs):
    max_state = 0
    max_reward = -10000
    for i in range(n_states):
      for a in range(n_actions):
        sum_rewards = 0
        for j in range(n_states):
          sum_rewards = sum_rewards + P[i][a][j]*(R[i][a][j] + gamma*V_old[0, j])
        if (sum_rewards >= max_reward):
          max_reward = sum_rewards
          max_state = a
      V_new[0, i] = max_reward
      if (e%steps_remember == 0):
        V[i].append(max_reward)
        A[i].append(max_state)
    increment = 0
    for i in range(n_states):
      increment = increment + (V_new[0, i] - V_old[0, i])**2
    if (np.sqrt(increment) < threshold):
      break
    V_old = V_new
    V_new = np.zeros([1, n_states])
  return(V,A)
```

```python
(V,A) = trainMDP(P, R, 0.8, 10, 0.001)
n_iterations = len(V[0])
n_states = len(V)
V_mat = np.zeros([n_states, n_iterations])
A_mat = np.zeros([n_states, n_iterations])
for i in range(n_states):
  for j in range(n_iterations):
    V_mat[i,j] = V[i][j]
    A_mat[i,j] = A[i][j]
```

```python
print("Maximum reward possible to gain for each state over time:")
print(V_mat)
print("\n")
print("Actions that maximize the reward for each state over time:")
print(A_mat)
```
